#  Plots- 2D and 3D, Graph Customization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xvalues = np.linspace(-np.pi, np.pi)
yvalues1 = np.sin(xvalues)
yvalues2 = np.cos(xvalues)
plt.plot(xvalues, yvalues1, lw=2, color='red',
label='sin(x)')
plt.plot(xvalues, yvalues2, lw=2, color='blue',
label='cos(x)')
plt.title('Trigonometric Functions')
plt.xlabel('x')
plt.ylabel('sin(x), cos(x)')
plt.axhline(0, lw=0.5, color='black')
plt.axvline(0, lw=0.5, color='black')
plt.legend()
None
```

```python
xvalues = np.linspace(-np.pi, np.pi)
yvalues1 = np.sin(xvalues)
yvalues2 = np.sin(2*xvalues)
yvalues3 = 2*np.sin(xvalues)
plt.plot(xvalues, yvalues1, lw=2, color='red',
label='sin(x)')
plt.plot(xvalues, yvalues2, lw=2, color='blue',
label='sin(2x)')
plt.plot(xvalues, yvalues3, lw=2, color='green',
label='2*sin(x)')
plt.title('Trigonometric Functions')
plt.xlabel('x')
plt.ylabel('sin(x), sin(2x) ,2*sin(x)')
plt.axhline(0, lw=0.5, color='black')
plt.axvline(0, lw=0.5, color='black')
plt.legend()
None

```

```python
plt.figure(figsize=(6,6))
xvalues = np.linspace(-2, 2, 100)
plt.subplot(2, 2, 1)
yvalues = xvalues
plt.plot(xvalues, yvalues, color='blue')
plt.xlabel('$x$')
plt.ylabel('$x$')
plt.subplot(2, 2, 2)
yvalues = xvalues ** 2
plt.plot(xvalues, yvalues, color='green')
plt.xlabel('$x$')
plt.ylabel('$x^2$')
plt.subplot(2, 2, 3)
yvalues = xvalues ** 3
plt.plot(xvalues, yvalues, color='red')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.subplot(2, 2, 4)
yvalues = xvalues ** 4
plt.plot(xvalues, yvalues, color='black')
plt.xlabel('$x$')
plt.ylabel('$x^4$')
plt.suptitle('Polynomial Functions')
plt.tight_layout()
plt.subplots_adjust(top=0.90)
None
```

```python
xx,zz=meshgrid(range(10),range(10)) 
y=0 
#plot the surface 
ax=axes(projection='3d') 
ax.plot_surface(xx,y,zz) 
title("Plane") 
ax.set_xlabel('$X$') 
ax.set_ylabel('$Y$') 
ax.set_zlabel('$Z$') 
show()
```

```python
ax=axes(projection='3d')
u=linspace(0,2*pi,100)
v=linspace(0,pi,100)
x=10*outer(cos(u),sin(v))
y=10*outer(sin(u),sin(v))
z=10*outer((ones(size(v))),cos(v))
ax.plot_surface(x,y,z,color='r')
show()
```

```python
## Vector field Plotting
from numpy import *
import matplotlib.pyplot as plt
```

```python


ax=axes(projection='3d')
x,y,z=meshgrid(arange(-0.8,1,0.2),
              arange(-0.8,1,0.2),
              arange(-0.8,1,0.8))
u=sin(pi*x)*cos(pi*y)*cos(pi*z)
v=-cos(pi*x)*sin(pi*y)*cos(pi*z)
w=(sqrt(2.0/3.0)*cos(pi*x)*cos(pi*y)*sin(pi*z))
ax.quiver(x,y,z,u,v,w,length=0.3)
show()
```

```python
x,y = meshgrid(linspace(-5,5,10),linspace(-5,5,10))
u = x/sqrt(x**2 + y**2)
v = y/sqrt(x**2 + y**2)
plt.quiver(x,y,u,v)
plt.show()
```
