### Solution of initial value problems

##### 1. Solve the IVP $\dfrac{dy}{dx}=-0.3y, y(0)=5$

```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def model(y,x):       # define a function that returns dy/dx
    dydx=-0.3*y
    return dydx

y0=5                 # intial condition
x=arange(0,20,0.5)   # timepoints

y1=odeint(model,y0,x) # solve ODE

plt.plot(x,y1)
plt.show()
```

##### 2. Solve the IVP $\dfrac{dy}{dx}=2k, y(0)=1$

```python
def model(y,t,k):       # define a function that returns dy/dx
    dydx=2*k
    return dydx
y0=1                 # intial condition
t=linspace(0,20,50)   # timepoints
# solve ODE
k=1
yk1=odeint(model,y0,t,args=(k,)) 
k=2
yk2=odeint(model,y0,t,args=(k,)) 
k=5
yk3=odeint(model,y0,t,args=(k,)) 
plt.plot(t,yk1,'r')
plt.plot(t,yk2,'b.-')
plt.plot(t,yk3,'k+')
plt.show()
```

##### 3. Solve the IVP $\dfrac{dy}{dx}=1-y, y(0)=0$

```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def model(y,x):       # define a function that returns dy/dx
    dydx=1-y
    return dydx

y0=0                 # intial condition
x=linspace(0,20)  # timepoints

yx=odeint(model,y0,x) # solve ODE

plt.plot(x,yx,'r-.')
plt.show()
```

### Solving higher order differential equations
To solve higher order differential equations, we have to convert it to a system of first order differential equations using the following method and then solve by using $odeint$. <br>
$\textbf{Example}$: Solve $y''+2y'+y=0$ ; ------------------------------------------ (1) 
   with intial conditions $[y(0)=5, y'(0)=3]$ <br> 
   
Here, take $y'=t$             ----------------------------------------- (2) <br> 
      Substitute (2) in (1), we get $t'+2t+y=0$ <br>
      i.e. $t'=-2t-y$ ----------------------------------------------(3) <br> 
      
Equations (2) and (3) constitute the system of first order differential equations. <br>

Now, let $u$ be a vector, $u=[y,t]$. <br> 
This implies $u[0]=y$ and $u[1]=t$. <br><br>
Then, (2) and (3) can be written as: <br>
      $y'=u[1]$ <br>
      $t'=-2u[1]-u[0]$ <br> ------------------------------------------------------------------------------ 
      
$\textbf{Step 1}$ : Define a function which returns the system of differential equations. <br>
$\textbf{Step 2}$ : Assign the initial conditions and define a range for independent variable x. <br>
$\textbf{Step 3}$ : Using in built function $\textit{odeint}$ solve the D.Eqns. <br> Syntax - $us=odeint(func,IC,x)$ <br> 
$\textbf{Step 4}$ : The solution $us$ will be an array with 2 columns.<br> First column for $u[0]=y$ and second column for $u[1]=t$. (With number of rows depending on the number of values defined for x).<br>
Therefore, $Sol=us[:,0]$ will extract the values obtained for $y$ from us. <br>
$\textbf{Step 5}$ : Plot the solution for Sol. 

##### 1. Solve the differential equation $y''+2y'+y=0$ with the initial conditions, $y(0)=5, y'(0)=-3$.

```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def du_dx(u,x):
    return [u[1],-2*u[1]-u[0]]
u0=[5,-3]
xs=linspace(0,10,200)
us=odeint(du_dx,u0,xs)
ys=us[:,0]
plt.plot(xs,ys,color="black")
plt.title("Solution of $y''+2y'+y=0$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

##### 2. Solve the differential equation $y''+2y'+2y=cos(2x)$ with the initial conditions, $y(0)=0, y'(0)=0$.

```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def du_dx(u,x):
    return [u[1],-2*u[1]-2*u[0]+cos(2*x)]
u0=[0,0]
xs=linspace(0,10,200)
us=odeint(du_dx,u0,xs)
ys=us[:,0]
plt.plot(xs,ys,color="red")
plt.title("Solution of $y''=2y'+2y=cos(2x)$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

##### 3. Solve the differential equation $y''+6y'+13y=0$ with the initial conditions, $y(0)=2, y'(0)=1$.

```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def du_dx(u,x):
    return [u[1],-6*u[1]-13*u[0]]
u0=[2,1]
xs=linspace(0,10,200)
us=odeint(du_dx,u0,xs)
ys=us[:,0]
plt.plot(xs,ys,color="red")
plt.title("Solution of $y''+6y'+13y=0$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

##### 4. Solve the differential equation $y''+3y'-5y=0$ with the initial conditions, $y(0)=0, y'(0)=5$.

```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def du_dx(u,x): # defining a function to return the differential equation
    return [u[1],-3*u[1]+5*u[0]] 
u0=[0,5]  # initial conditions                        
xs=linspace(0,20,200)  #timepoints (x-axis)
us=odeint(du_dx,u0,xs) #solving the d.e to obtain solution array 'us'
# Array 'us' has two columns. The first column gives solution 'y'.
# Each row (r) contains value of y for a timepoint, t=r.
ys=us[:,0] #Extracting the first column (Solution set - y)
plt.plot(xs,ys,color="black") # Plotting the solution-set(yaxis) against timepoints(xaxis)
plt.title("Solution of $y''-y'+9y=0$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
############
print(us[5,0],us[10,0],us[15,0])
# To extract the solution y for a timepoint t=5(say), then you may use the code us[5,0].
```
