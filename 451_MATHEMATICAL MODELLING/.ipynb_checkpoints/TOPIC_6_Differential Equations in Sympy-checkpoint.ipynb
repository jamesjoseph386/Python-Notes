### 1) solve the differential equation dy/dx = 1

```python

from sympy import Symbol, Function
from sympy.plotting import plot
x=Symbol("x")#defining independant
y=Function("y")(x)#defining Dependant
D=Eq(y.diff(x),1)#assignign differential equation
print(D)
display(D)
print(dsolve(D,y)

plot(x, (x,-5,5))
```

### 2)Solve: dp/dt − r = 0

```python

from sympy import*
t=Symbol("t")#defining independant
p=Function("p")(t)#defining Dependant
r=Symbol("r")
D=Eq(p.diff(t,1)-r, 0)#assignign differential equation
print(D)
display(D)
dsolve(D,p)
```

### 3)Solve: 6y + 5dy/dx + d^2 y/dx^2 = 0

```python

from sympy import*
x=Symbol("x")#defining independant
y=Function("y")(x)#defining Dependant
D=Eq(y.diff(x,2)+5*y.diff(x)+6*y,0)#assigning differential equation
print(D)
display(D)
dsolve(D,y)
```

### 4)Solve: d^3 y/dx^3 + y = 0

### 5)Solve: d^3 y/dx^3 + y = 0; y(2) = 2, y(0) = 1, y(−1) = 0

```python

from sympy import*
x=Symbol("x")#defining independant
y=Function("y")(x)#defining Dependant
D=Eq(y.diff(x,3)+y,0)#assigning differential equation
print("Differential equation:")
display(D)
Gs1=dsolve(D,y)
print("General solution:")
display(Gs1)
print("Particular solutions:")
S1=Gs1.subs({y:2,x:2})
display(S1)
S2=Gs1.subs({y:1,x:0})
display(S2)
S3=Gs1.subs({y:0,x:-1})
display(S3)
```

### 6) Find the general solution for the given DE: dp/dt - sin(t) - cos(t) and further

### 7) Solve : df/dx = 5f; f(2) = 10.

```python


from sympy import *
x=Symbol("x")
f=Function("f")(x)
print("Differential equation")
diff1=Eq(f.diff(x),5*f)
display(diff1)
print("General Solution")
Gs1=dsolve(diff1)
display(Gs1)
print("Particular solution is ")
ps1=Gs1.subs({x:2,f:10})
display(ps1)
```

### Solve the following initial value problems

### 1)  $(y+2)dx + y(x+4)dy = 0; y(-3) = -1.$


```python
from sympy import *


x=Symbol("x")
y=Function("y")(x)

print("Differential equation")
diff1=Eq(y.diff(x),(-y-2)/(y*(x+4)))
display(diff1)

print("General Solution")
Gs=dsolve(diff1)
display(Gs)

C1=Symbol("C1")

ps0=Gs[0].subs({x:-3,y:-1})
display(ps0)
c0=solve(ps0,C1)
display(c0)

ps1=Gs[1].subs({x:-3,y:-1})
display(ps1)
c1=solve(ps1,C1)
display(c1)

print("Particular solution is ")
p0 = Gs[0].subs({C1:c0[0]})
display(p0)

```

### 2) $x sin(y) dx + (x^2 + 1)cos(y)dy = 0; y(1) = \frac{\pi}{2}.$

```python
from sympy import *


x=Symbol("x")
y=Function("y")(x)

print("Differential equation")
diff1=Eq(y.diff(x),(-x*sin(y))/((x**2+1)*cos(y)))
display(diff1)

print("General Solution")
Gs1=dsolve(diff1)
display(Gs1)

C1=Symbol("C1")

ps0=Gs1[0].subs({x:1,y:(pi/2)})
display(ps0)
c0=solve(ps0, C1)
display(c0)

ps1=Gs1[1].subs({x:1,y:(pi/2)})
display(ps1)
c1=solve(ps1, C1)
display(c1)

print("Particular solution is ")
p0 = Gs1[0].subs({C1:c0[0]})
display(p0)

p1 = Gs1[1].subs({C1:c1[0]})
display(p1)
```


### 3) $8 cos^2(y) dx + cosec^2(x)dy = 0; y(\frac{\pi}{12}) = \frac{\pi}{4}.$

```python
from sympy import *

x=Symbol("x")
y=Function("y")(x)

print("Differential equation")
diff1=Eq(y.diff(x),-8*cos(y)**2/csc(y)**2)
display(diff1)

print("General Solution")
Gs1=dsolve(diff1)
display(Gs1)


ps1=Gs1.subs({x:pi/12,y:pi/4})
display(ps1)
C1=Symbol("C1")
c1=solve(ps1, C1)
display(c1)

print("Particular solution is ")
ps = Gs1.subs({C1:c1[0]})
display(ps)
```


### 4) $(3x + 8)(y^2 + 4)dx - 4y(x^2 + 5x + 6)dy = 0; y(1) = 2.$

```python
from sympy import *

x=Symbol("x")
y=Function("y")(x)

print("Differential equation")
diff1=Eq(y.diff(x),(3*x+8)*(y**2+4)/(4*y*(x**2+5*x+6)))
display(diff1)

print("General Solution")
Gs1=dsolve(diff1)
print(Gs1)

C1=Symbol("C1")


ps0=Gs1[0].subs({x:1,y:2})
display(ps0)
c0=solve(ps0, C1)
display(c0)

ps1=Gs1[1].subs({x:1,y:2})
display(ps1)
c1=solve(ps1, C1)
display(c1)

print("Particular solution is ")

ps10 = Gs1[1].subs({C1:c1[0]})
display(ps10)

ps11 = Gs1[1].subs({C1:c1[1]})
display(ps11)
```

### 5) $(x^2 + 3y^2)dx - 2xydy = 0; y(2) = 6.$

```python
from sympy import *
x = Symbol("x")
y = Function("y")(x)

deq = Eq(y.diff(x), (x**2 + 3*y**2)/(2*x*y))
display(deq)
```

```python
gs = dsolve(deq,y)
display(gs)
```

```python
C1 = Symbol("C1")
```

```python
ps0 = gs[0].subs({x:2,y:6})
display(ps0)
c0=solve(ps0,C1)
display(c0)
```

```python
ps1 = gs[1].subs({x:2,y:6})
display(ps1)
c1=solve(ps1,C1)
display(c1)
```

```python
print("The particular solution is:")
p = gs[1].subs({C1:c1[0]})
display(p)
```
