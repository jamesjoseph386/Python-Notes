6. Write a Python program to slice a tuple which shows: 
i. only items starting from 0 upto 3
ii. only items starting from 4 to the end
iii. only items starting from 2 upto 6

```python
tup = ('abc', 'def', 'ghi', 'jklm', 'nopqr', 'st', 'uv', 'wxyz', '23', 's98', '123', '87')
print("The given tuple is:",tup)
# Slicing
# shows only items starting from 0 upto 3
print('tuple showing only items starting from index 0 upto 3\n', tup[:3])

# shows only items starting from 4 to the end
print('tuple showing only items starting from  index 4 to the end\n', tup[4:])

# shows only items starting from 2 upto 6
print('tuple showing only items starting from index 2 upto 6\n', tup[2:6])

```
