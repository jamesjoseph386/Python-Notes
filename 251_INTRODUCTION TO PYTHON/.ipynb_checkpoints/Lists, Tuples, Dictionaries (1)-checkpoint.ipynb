```python
#Making a list of marks and finding the percentages
marks=[[94,96,84,92],[86,89,94,90]]
#print(marks)
per=[]
n=2
for i in range(n):
    p=(sum(marks[i]))/4
    per.append(p)
#print(per)
#print("Percentage of marks obtained by Bob: ",per[0])
#print("Percentage of marks obtained by Ken: ",per[1])

```

```python
#Sample database without the subject names
database={'Bob':{'Register Number':'1840083','Marks': marks[0],'Percentage': per[0]},
          'Ken':{'Register Number':'2040086','Marks': marks[1],'Percentage': per[1]}}
```

```python
#Sample database with the subject names
database={'Bob':{'Register Number':'1840083','Marks': marks[0],'Percentage': per[0]},
          'Ken':{'Register Number':'2040086',
                    'Marks':{'Sc':marks[1][0],'Maths':marks[1][1],'Eng':marks[1][2],
                             'Elective':marks[1][3]},'Percentage': per[1]}}
```
