1. Write a Python program to add an item in a specific index of a tuple without converting it into a list.

```python

tup=eval(input("Enter a tuple of 10 elements"))
new_tup=tup[:5] + (2,3) +tup[5:]
print(new_tup)

```

2. Write a Python program to find the number of times an item is repeated in a tuple

```python

tup=eval(input("Enter a tuple where some elements are repeated"))
rep=int(input("Enter the value to be checked for repetition"))
count=tup.count(rep)
print(count)
```

3. Write a Python program to remove an item from a tuple without converting it into a list.

```python

tup=eval(input("Enter a tuple"))
rem=int(input("Enter the index of the item to be removed"))
new_tup=tup[:rem]+tup[rem+1:]
print(new_tup)
```

4. Write a Python program to compute the sum of all the elements of each tuple stored inside
a list of tuples.

```python

tup=eval(input("Enter a list of tuples"))
result = [sum(t) for t in tup]
print(result)
```

5. Write a Python program to convert a given list of tuples to a list of lists.

```python

tup= eval(input("Enter a list of tuples"))
result = [list(el) for el in tup]
print(result)
```

6. Write a program to find the sum of all the elements in a dictionary 

```python
def returnSum(myDict):
     
    list = []
    for i in myDict:
        list.append(myDict[i])
    final = sum(list)
     
    return final
 
# Driver Function
dict = eval(input("Enter a dictionary of numbers:"))
print("Sum :", returnSum(dict))
```

7. Write a python program to change the elements of the list ["Introduction", "to", "Python"] to the corresponding Kth element
of the list in the dictionary {"Introduction" : [5, 6, 7], "Python" : [7, 4, 2]}.

```python
# initializing list
l = ["Introduction", "to", "Python"]
  
# printing original list
print("The original list : ", l)
  
# initializing Dictionary
d = {"Introduction" : [5, 6, 7], "Python" : [7, 4, 2], }
  
# initializing K 
K = 2
  
# using list comprehension to solve
# problem using one liner
res = [ele if ele not in d else d[ele][K]
                                     for ele in l]
          
# printing result 
print("The list after substitution : " res)
```

8. Write a python program to input a list of numbers with repetion and create a dictionary with the numbers of the list as 
the keys and their frequency in the list as the values.

```python

def CountFrequency(lst):
 
    # Creating an empty dictionary
    freq = {}
    for item in lst:
        if (item in freq):
            freq[item] += 1
        else:
            freq[item] = 1
 
    for key, value in freq.items():
        print ("% d : % d"%(key, value))
 

l = eval(input(" Enter a list of numbers with repetition"))
 
CountFrequency(l)
```
