```python
A = set(range(1,11))
print("A =",A)
B = set(range(5,16))
print("B =",B)
U = set(range(1,26))
print("U =",U)
v=venn2(subsets=(4,5,6),set_labels=("A","B"))
```

```python
i= A.intersection(B)
print("1. A∪B has {0} elements given by {1}".format(len(i),i))
u= A.union(B)
print("2. A∩B has {0} elements given by {1}".format(len(u),u))
x= A.difference(B)
print("3. A-B has {0} elements given by {1}".format(len(x),x))
x= B.difference(A)
print("4. B-A has {0} elements given by {1}".format(len(x),x))
p=(U-A).union(U-B)
print("5. A'∪B' has {0} elements given by {1}".format(len(p),p))
q=U-i
print("6. (A∩B)' has {0} elements given by {1}".format(len(q),q))
print("\n Since A'∪B'=(A∩B)' \n De Morgan's law is verified.")
```
