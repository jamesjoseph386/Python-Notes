1. A student has 10 questions. Assuming all the 10 attempts are correct, the marks are
evaluated based on the types of questions attempted. The details of the questions are given below:

Question Type Marks

A 2

B 3

C 4

D 5

If the num of questions attempted of type A are more than 8, the final marks are incremented by 4%.

If the num of questions attempted of type B are more than 7, the final marks are
incremented by 5%.

If the num of questions attempted of type C are more than 6, the final marks are
incremented by 3%.

If the num of questions attempted of type D are more than 5, the final marks are
incremented by 1%.

Evaluate the final marks for the following cases:

```python
a=3
b=0
c=1
d=6
f=0
tot=0
n = [a,b,c,d]
if n[0]>8:
  f = f+4
if n[1]>7:
  f = f+5
if n[2]>6:
  f = f+3
if n[3]>5:
  f = f+1
for i in range(0,len(n)):
  if i==0:
    tot= tot+2*n[i]
  if i==1:
    tot= tot+3*n[i]
  if i==2:
    tot= tot+4*n[i]
  if i==3:
    tot= tot+5*n[i]
print("Total Marks:",tot+(f/100.0)*tot)
```
