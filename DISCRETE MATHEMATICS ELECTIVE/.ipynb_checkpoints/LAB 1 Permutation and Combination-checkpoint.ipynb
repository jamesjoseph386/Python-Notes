# PERMUTATIONS AND COMBINATIONS

Permutation is an arrangement of objects in a specific order. The number of permutations on a set of n elements is given by  n!.
A permutation, also called an “arrangement number” or “order,” is a rearrangement of the elements of an ordered list S into a one-to-one correspondence with S itself.

Combination is a collection of the element where the order doesn't matter. Python itertools module provides the combination() method to calculate the combination of given data.

Q. Write a program to calculate the following:

(a) How many words can be formed by using 3 letters from the word “DELHI”?


(b) In how many ways, can we select a team of 4 students from a given choice
of 15 ?

```python
import math
def fact(n):
    if n==1 or n==0:
        return 1
    else:
        return n*fact(n-1)

print("5!=",fact(5))
```

```python
# Combination:
def Comb(n,r):
  if n>=r:
    return math.factorial(n)/(math.factorial(n-r)*math.factorial(r))
  else:
    print("Invalid Entry")
```

```python
def Perm(n,r):
    if n>=r:
        return fact(n)/(fact(n-r))
    else:
        print("Invalid Entry")
```

```python
#a)
# Number of letters in the word "DELHI" = 5
# Number of letters in words to be made = 3
# Number of 3 letter words that can be made from the word "DELHI" = 5 P 3
print("Number of 3 letter words that can be made from the word 'DELHI' is:",Perm(5,3))
```

```python
#b)
# Total number of students = 15
# Number of students to be selected = 4
# Number of combinations of 4-team students from a choice of 15 = 15 C 4
print("Number of ways we can select a team of 4 students from a chioce of 15 is:",Comb(15,4))
```

Q. Find the permutations of 1,2,3.

```python
# A Python program to print all permutations using library function
from itertools import permutations  #itertools - is used to iterate over data structures that can be stepped over using a for-loop.

# Get all permutations of [1, 2, 3]
perm = permutations([1, 2, 3])
# Print the obtained permutations
for i in list(perm):
    print (i)
```

1. Combinations are emitted in lexicographic sort order of input. So, if the input list is sorted, the combination tuples will be produced in sorted order.

```python
# A Python program to print all
# combinations of a given length
from itertools import combinations

# Get all combinations of [1, 2, 3]
# and length 2
comb = combinations([1, 2, 3], 2)

# Print the obtained combinations
for i in list(comb):
	print (i)
```

2. Elements are treated as unique based on their position, not on their value. So if the input elements are unique, there will be no repeat values in each combination.

```python
# A Python program to print all combinations
# of given length with unsorted input.
from itertools import combinations

# Get all combinations of [2, 1, 3]
# and length 2
comb = combinations([2, 1, 3], 2)

# Print the obtained combinations
for i in list(comb):
	print (i)
```

Q. Write a function to evaluate the following
1. Number of arrangements of n objects taking r at a time
2. Number of selections of n objects taking r at a time

```python
def perm(n,r):
    num = 1
    den = 1
    for i in range (1,n+1):
        num = i*num
    for j in range (1,n-r+1):
        den = j*den
    return num/den

def comb(n,r):
    num = 1
    den = 1
    den1 = 1
    for i in range (1,n+1):
        num = i*num
    for j in range (1,n-r+1):
        den = j*den
    for k in range (1,r+1):
        den1 = k*den1
    return num/(den*den1)
```

Q) Build 5 rows of Pascal Triangle.

```python
import numpy as np

r = int(input("Enter num of rows for the Pascal Traingle"))
n= np.zeros((r,(r-1)*2+1),int)
#Filling with ones
for i in range(0,np.shape(n)[0]):
  for k in range(0,np.shape(n)[1]):
    if k==np.shape(n)[1]//2:
      try:
        n[i,k-i]=1
        n[i,k+i]=1
      except:
        pass
#To add all remaining nums
for i in range(0,np.shape(n)[0]):
  for k in range(0,np.shape(n)[1]):
    try:
      if n[i-1,k-1]!=0 and n[i-1,k+1]:
        n[i,k]=n[i-1,k-1]+n[i-1,k+1]
    except:
      pass

print(n)
#To print with spaces
for i in range(0,np.shape(n)[0]):
  for k in range(0,np.shape(n)[1]):
    if n[i,k]==0:
      print(' ',end='')
    else:
      print(n[i,k],end='')
  print()
```
