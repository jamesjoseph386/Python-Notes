### During freshman orientation in a small liberal arts college 2 showings of a latest James bond movie were presented among the 600 freshmen, 80 attended the first showing and 125 attended the second showing while 450 didn't make it to either of the showing. How many of 600 freshmen attended twice?

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
venn2(subsets=(80-((80+125)-150),125-((80+125)-150),(80+125)-150),set_labels=('A','B'))
#plt.gca().set_axis_bgcolor('skyblue')
#plt.gca().set_axis.on()
plt.show()
```

### A manufacturer of 2000 automobile batteries is concerned anout defective terminals and defective plates, if1920 off her batteries have neither defect, 60 have defective plates and 20 have both defects. How many batteries have defective terminals?

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
venn2(subsets = (40,20,20), set_labels = ('Defective Plates', 'Defective Terminals'))
#plt.gca().set_axis_bgcolor('lightgrey')
#plt.gca().set_axis_on()
plt.show()
```

### A binary string is made up of 12 bits i.e., 2 symbols each of which is 0 or 1, how many such strings either start with three 1's or end in four 0's.

```python
print("The number of strings are:", 2**9 + 2**8)
```

Determine |AvBvC| when |A| = 50 |B| = 500 and |C| = 5000  
If:
1. A<=B<=C
2. A^B = A^C = B^C = null
3.|A^B| = |A^C| = |B^C| = 3 and |A^B^C| = 1

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from matplotlib_venn import venn3
print("|AvBvC| = ",5000)
venn3(subsets = (0,0,0,4500,0,450,50), set_labels = ('A', 'B', 'C'))
plt.show()
```

```python
# Here A intersection B=A intersection C=B intersection C=empty
import matplotlib.pyplot as plt
import numpy
from matplotlib_venn import venn2
from matplotlib_venn import venn3
print("|AvBvC| = ",5550)
venn3(subsets = (50,500,0,5000,0,0,0), set_labels = ('A', 'B', 'C'))
plt.show()
```

```python
# Here cardinality of A intersection B=cardinality of A intersection C=cardinality
#AND cardinality of A intersection B intersection C=1
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from matplotlib_venn import venn3
print("|AvBvC| = ",5542)
venn3(subsets = (50-5,500-5,2,5000-5,2,2,1), set_labels = ('A', 'B', 'C'))
plt.show()
```
