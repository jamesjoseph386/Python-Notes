## Vector Plots

```python
from numpy import *
import matplotlib.pyplot as plt
```

### 1) Write a program to obtain a vector plot for
### $u=\frac{x}{x^2+y^2}$, $v=\frac{y}{x^2+y^2}$, $w=\frac{x^2}{x^2+y^2}-\frac{y^2}{x^2+y^2}$

```python
fig=figure()
ax=fig.gca(projection='3d')
x,y,z=meshgrid(arange(-0.8,1,0.2),
              arange(-0.8,1,0.2),
              arange(-0.8,1,0.8))
u=(x)/(x**2+y**2)
v=(y)/(x**2+y**2)
w=(x**2)/(x**2+y**2)-(y**2)/(x**2+y**2)
ax.quiver(x,y,z,u,v,w,length=0.1)
show()
```

### Write a program to obtain a vector plot of $u=\frac{x}{\sqrt{x^2+y^2}}$ and $v=\frac{y}{x^2+y^2}$

```python
x,y = meshgrid(linspace(-5,5,10),linspace(-5,5,10))
u = x/sqrt(x**2 + y**2)
v = y/sqrt(x**2 + y**2)
plt.quiver(x,y,u,v)
plt.show()
```

## Stream Plot

### Write a program to obtain a stream plot of $u=\frac{-y}{\sqrt{x^2+y^2}}$ and $v=\frac{x}{x^2+y^2}$

```python
x,y = meshgrid(linspace(-5,5,10),linspace(-5,5,10))
u = -y/(x**2 + y**2)
v = x/(x**2 + y**2)
plt.streamplot(x,y,u,v,color= 'Red',density=1,cmap='copper',arrowsize=1)
plt.show()
```
