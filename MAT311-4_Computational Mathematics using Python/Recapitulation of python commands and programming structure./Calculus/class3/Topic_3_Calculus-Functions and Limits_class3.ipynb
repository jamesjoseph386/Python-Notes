# Limits

We can find limits of functions in SymPy by creating objects of the Limit
class as follows

```python
#  S, which is a special SymPy class that contains the definition of infinity (positive and negative) and other special values.
from sympy import Limit, Symbol, S
x = Symbol('x')
m = Limit(1/x, x, S.Infinity)
m
```

```python
#To find the value of the limit, we use the doit() method
# doit method is used to evaluate the function which is not evaluated by default
m.doit()
```

To determine value from LHS and RHS

1) $ f(x) = x^2 + 2x +1$

```python
import numpy as np
def limit():
    n=float(input("Enter the value of limit point: "))
    print("z tends to n from left hand side")
    l=np.linspace(n-0.1,n,10)
    f1=l**2+2*l+1
    print(f1)
    print("z tends to n from right hand side")
    m=np.linspace(n,n+0.1,10)
    f2=m**2+2*m+1
    print(f2[::-1])
    print("z at n")
    f3=n**2+2*n+1
    print(f3)
```

```python
limit()
```

2) $ f(x) = 2x + 4$

```python
def limit():
    n=float(input("Enter the value of limit point: "))
    print("z tends to n from left hand side")
    l=np.linspace(n-0.1,n,10)
    f1= 2*l + 4
    print(f1)
    print("z tends to n from right hand side")
    m=np.linspace(n,n+0.1,10)
    f2= 2*m + 4
    print(f2[::-1])
    print("z at n")
    f3= 2 *n +4
    print(f3)
    
```

```python
limit()
```
