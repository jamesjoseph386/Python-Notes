# Check the commutativity for addition and multiplication of matrices and display the result.

$A=\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9 \\
\end{bmatrix}$

$B=\begin{bmatrix}
-1 & 4 & 0 \\
2 & 3 & 1 \\
2 & 1 & 0 \\
\end{bmatrix}$

```python
A = np.matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
B = np.matrix([[-1, 4, 0], [2, 3, 1], [2, 1, 0]])
```

```python
# Addition
print("A =", A)
print("B =", B)
print("A+B =", A+B)
print("B+A =", B+A)
if np.allclose(A+B, B+A):
    print("A+B = B+A")
    print("Hence, matrix addition is commutative")
else:
    print("A+B != B+A")
    print("Hence, matrix addition is not commutative")
```

```python
# Multiplication
print("A =", A)
print("B =", B)
print("AB = ", A*B)
print("BA = ", B*A)
if np.allclose(A*B, B*A):
    print("AB = BA")
    print("Hence, matrix multiplication is commutative")
else:
    print("AB != BA")
    print("Hence, matrix multiplication is not commutative")
```

# Find $A^3 - 3A + 20I$

```python
A = np.matrix([[1, 2, 3, 4], [5, 6, 7, 8], [9, 8, 7, 6], [5, 4, 3, 2]])
eq = A**3 - 3*A + 20*np.identity(4)
print("A^3 - 3A + 20I =\n", eq)
```

Express the Matrix
$$\begin{bmatrix} 2 & 3 & 1 \\ 4 & -2 & 0 \\ 13 & -1 & 5\end{bmatrix}$$
as sum of symmetric and skew symmetric matrices.</b>

```python
A=np.array([[2,3,1],[4,-2,0],[3,-1,5]])
B=np.identity(3,dtype=int)
print("Matrix \n",A)
T=np.transpose(A)
print("\nTranspose of A \n",T)
sol1=(A+T)/2
print("Symmetric matrix")
print("\n(A + A^T)/2= \n",sol1)
sol2=(A-T)/2
print("Skew Symmetric matrix.")
print("\n(A - A^T)/2= \n",sol2)
ans=sol1+sol2
print("\nAnswer = \n",ans)
print("\nTherefore, A = sum of symmetric and skew symmetric matrix")
```
