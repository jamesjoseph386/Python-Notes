# Introduction to Python

# Variables and Data Types
Python supports various data types such as integers, floats, strings, lists, and dictionarise


```python
# Integer
x = 10
print("Integer:", x)

# Float
y = 3.14
print("Float:", y)

# String
name = "Alice"
print("String:", name)

# Boolean
is_valid = True
print("Boolean:", is_valid)
```

# Lists
Lists are ordered, mutable collections of items.

```python
# Creating a list
fruits = ["apple", "banana", "cherry"]

# Accessing items
print("First fruit:", fruits[0])

# Adding an item
fruits.append("date")
print("List after adding an item:", fruits)

# Removing an item
fruits.remove("banana")
print("List after removing an item:", fruits)

# Slicing
print("First two fruits:", fruits[:2])
```

# Dictionaries
Dictionaries store data in key-value pairs


```python
# Creating a dictionary
person = {"name": "Alice", "age": 25, "city": "New York"}

# Accessing values
print("Name:", person["name"])

# Adding a key-value pair
person["job"] = "Engineer"
print("Updated dictionary:", person)

# Removing a key-value pair
del person["age"]
print("Dictionary after deletion:", person)
```

# Loops
Loops allow you to iterate over items in a collection or execute code multiple times


```python
# For loop
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print("Fruit:", fruit)

# While loop
count = 0
while count < 3:
    print("Count:", count)
    count += 1
```

# Conditional Statements
Use `if`, `elif`, and `else` to control the flow of the program based on conditions.

```python
x = 10
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")
```

# Functions
Functions allow you to encapsulate reusable code


```python
# Defining a function
def greet(name):
    return f"Hello, {name}!"

# Calling a function
message = greet("Alice")
print(message)
```

# List Comprehensions
A concise way to create lists


```python
# Create a list of squares
squares = [x**2 for x in range(5)]
print("Squares:", squares)

# Filter even numbers
even_numbers = [x for x in range(10) if x % 2 == 0]
print("Even numbers:", even_numbers)
```

# Exception Handling
Use `try`, `except`, and `finally` to handle errors.

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")
finally:
    print("Execution complete.")
```

# Data Manipulation with Pandas

```python
import seaborn as sns
import pandas as pd

# Load Titanic dataset from Seaborn
df = sns.load_dataset('titanic')

# Display the first few rows of the dataset
print("First 5 rows of the Titanic dataset:")
print(df.head())

# Dataset information
print("\nDataset Information:")
print(df.info())
```

```python
# Display basic statistics
print("Summary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Fill missing values in the 'age' column with the median
df['age'] = df['age'].fillna(df['age'].median())

# Check if 'cabin' exists before trying to drop rows
if 'cabin' in df.columns:
    df = df.dropna(subset=['cabin'])
else:
    print("Column 'cabin' does not exist. Skipping this step.")

# Renaming columns
df.rename(columns={'fare': 'Ticket Price'}, inplace=True)

print("\nCleaned Dataset:")
print(df.head())
```

```python
print(df.columns)
print(df.isnull().sum())
```

```python
# Sort by age in ascending order
sorted_by_age = df.sort_values(by='age', ascending=True)
print("\nDataset sorted by age:")
print(sorted_by_age.head())

# Filter passengers who paid more than 50 for their ticket
high_fare_passengers = df[df['Ticket Price'] > 50]
print("\nPassengers who paid more than 50:")
print(high_fare_passengers.head())
```

```python
# Add a column for family size
df['family_size'] = df['sibsp'] + df['parch'] + 1
print("\nDataset with family size:")
print(df[['sibsp', 'parch', 'family_size']].head())

# Modify an existing column: categorize age
df['age_group'] = df['age'].apply(lambda x: 'Child' if x < 18 else 'Adult')
print("\nDataset with age groups:")
print(df[['age', 'age_group']].head())
```

```python
# Group by class and calculate average ticket price
avg_ticket_price = df.groupby('pclass')['Ticket Price'].mean()
print("\nAverage ticket price by class:")
print(avg_ticket_price)

# Group by gender and calculate survival rate
survival_rate = df.groupby('sex')['survived'].mean()
print("\nSurvival rate by gender:")
print(survival_rate)
```

```python
# Pivot table for survival rates by class and gender
pivot_table = pd.pivot_table(df, values='survived', index='pclass', columns='sex', aggfunc='mean')
print("\nSurvival rates by class and gender:")
print(pivot_table)
```

```python
# Average age by class
avg_age_by_class = df.groupby('pclass')['age'].mean()
print("\nAverage age by class:")
print(avg_age_by_class)

# Total fare by embarkation point
total_fare_by_embark = df.groupby('embarked')['Ticket Price'].sum()
print("\nTotal fare collected by embarkation point:")
print(total_fare_by_embark)

# Categorize survival status
df['survival_status'] = df['survived'].apply(lambda x: 'Survived' if x == 1 else 'Did Not Survive')
print("\nDataset with survival status:")
print(df[['survived', 'survival_status']].head())
```

```python
import matplotlib.pyplot as plt

# Line plot: Age distribution
plt.plot(df['age'].sort_values().reset_index(drop=True))
plt.title('Age Distribution')
plt.xlabel('Index')
plt.ylabel('Age')
plt.show()

# Bar plot: Count of passengers by class
df['pclass'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset from Seaborn
df = sns.load_dataset('titanic')

# Drop rows where key values are missing
df = df.dropna(subset=['fare', 'age'])

# Convert 'pclass' to categorical type
df['pclass'] = df['pclass'].astype('category')

# Scatter Plot: Fare vs Age by Gender
sns.scatterplot(x='fare', y='age', hue='sex', data=df)
plt.title('Fare vs Age by Gender')
plt.show()

# Box Plot: Fare by Class
sns.boxplot(x='pclass', y='fare', data=df)
plt.title('Box Plot of Fare by Class')
plt.show()

# Heatmap: Correlation Matrix
numeric_df = df.select_dtypes(include=['number']).dropna()
correlation = numeric_df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset from Seaborn
df = sns.load_dataset('titanic')

# Drop rows where key values are missing
df = df.dropna(subset=['fare', 'age'])

# Convert 'pclass' to categorical type
df['pclass'] = df['pclass'].astype('category')

# Scatter Plot: Fare vs Age by Gender
sns.scatterplot(x='fare', y='age', hue='sex', data=df)
plt.title('Fare vs Age by Gender')
plt.show()

# Box Plot: Fare by Class
sns.boxplot(x='pclass', y='fare', data=df)
plt.title('Box Plot of Fare by Class')
plt.show()

# Heatmap: Correlation Matrix
numeric_df = df.select_dtypes(include=['number']).dropna()
correlation = numeric_df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Count number of passengers in each class
passenger_count = df['pclass'].value_counts()
print("\nNumber of passengers in each class:")
print(passenger_count)
```

```python
# Ensure there are no NaN values in the 'age' column before finding min/max
df_cleaned = df.dropna(subset=['age'])

# Find the oldest and youngest passenger
oldest_passenger = df_cleaned.loc[df_cleaned['age'].idxmax()]
youngest_passenger = df_cleaned.loc[df_cleaned['age'].idxmin()]

print("\nOldest Passenger Details:")
print(oldest_passenger[['who', 'age', 'pclass', 'sex']])  # No 'name' column in Seaborn's dataset

print("\nYoungest Passenger Details:")
print(youngest_passenger[['who', 'age', 'pclass', 'sex']])
```

```python
# Calculate survival rate for each class
survival_by_class = df.groupby('pclass')['survived'].mean() * 100
print("\nSurvival rate by class (percentage):")
print(survival_by_class)
```

```python
# Drop rows with missing 'fare' values to prevent errors
df_cleaned = df.dropna(subset=['fare'])

# Find the passenger who paid the highest fare
highest_fare_passenger = df_cleaned.loc[df_cleaned['fare'].idxmax()]

print("\nPassenger who paid the highest fare:")
# Adjust columns based on dataset availability
print(highest_fare_passenger[['who', 'fare', 'pclass', 'sex', 'embarked']])
```

```python
# Count number of survivors by gender
survival_by_gender = df.groupby('sex')['survived'].sum()
print("\nNumber of male vs female survivors:")
print(survival_by_gender)
```

```python
# Identify passengers who traveled alone
df['travel_alone'] = df.apply(lambda row: 'Yes' if row['sibsp'] + row['parch'] == 0 else 'No', axis=1)

# Count how many traveled alone
solo_travelers_count = df['travel_alone'].value_counts()
print("\nNumber of passengers who traveled alone:")
print(solo_travelers_count)
```

```python
# Drop NaN values in 'fare' before processing
df_cleaned = df.dropna(subset=['fare']).reset_index(drop=True)

# Find the passenger who paid the highest fare
highest_fare_index = df_cleaned['fare'].idxmax()
highest_fare_passenger = df_cleaned.loc[highest_fare_index]

print("\nPassenger who paid the highest fare:")
print(highest_fare_passenger[['who', 'fare', 'pclass', 'sex', 'embarked']])

# Find the passenger who paid the lowest non-zero fare
lowest_fare_index = df_cleaned[df_cleaned['fare'] > 0]['fare'].idxmin()
lowest_fare_passenger = df_cleaned.loc[lowest_fare_index]

print("\nPassenger who paid the lowest non-zero fare:")
print(lowest_fare_passenger[['who', 'fare', 'pclass', 'sex', 'embarked']])
```

```python
# Calculate average fare for survivors vs non-survivors
avg_fare_survivors = df[df['survived'] == 1]['fare'].mean()
avg_fare_non_survivors = df[df['survived'] == 0]['fare'].mean()

# Bar plot
sns.barplot(x=['Survivors', 'Non-Survivors'], y=[avg_fare_survivors, avg_fare_non_survivors], palette='pastel')
plt.title('Comparison of Average Fare: Survivors vs Non-Survivors')
plt.ylabel('Average Fare')
plt.show()
```
