```python
def input_matrix(name):
    print(f"Enter number of rows and columns for {name}:")
    rows = int(input("Rows: "))
    cols = int(input("Columns: "))

    print(f"Enter elements of {name} row-wise:")
    matrix = []
    for i in range(rows):
        row = list(map(int, input(f"Row {i+1}: ").split()))
        while len(row) != cols:
            print(f"Please enter exactly {cols} values.")
            row = list(map(int, input(f"Row {i+1}: ").split()))
        matrix.append(row)
    return matrix


def print_matrix(mat):
    for row in mat:
        print(*row)


def add_matrices(A, B):
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]


def subtract_matrices(A, B):
    return [[A[i][j] - B[i][j] for j in range(len(A[0]))] for i in range(len(A))]


def multiply_matrices(A, B):
    result = [[0] * len(B[0]) for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result


# --- MAIN PROGRAM ---
A = input_matrix("Matrix A")
B = input_matrix("Matrix B")

print("\nChoose operation:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")

choice = int(input("Enter your choice (1/2/3): "))

print("\nResult:")
if choice == 1:
    if len(A) == len(B) and len(A[0]) == len(B[0]):
        print_matrix(add_matrices(A, B))
    else:
        print("Addition not possible — matrix dimensions must match.")
elif choice == 2:
    if len(A) == len(B) and len(A[0]) == len(B[0]):
        print_matrix(subtract_matrices(A, B))
    else:
        print("Subtraction not possible — matrix dimensions must match.")
elif choice == 3:
    if len(A[0]) == len(B):
        print_matrix(multiply_matrices(A, B))
    else:
        print("Multiplication not possible — columns of A must equal rows of B.")
else:
    print("Invalid choice.")
    

```
