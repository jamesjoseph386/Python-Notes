```python
plainText = input("What is your plaintext? ")
shift = int(input("What is your shift? "))

def caesar(plainText, shift): 

    for ch in plainText:
        if ch.isalpha():
            stayInAlphabet = ord(ch) + shift 
            if stayInAlphabet > ord('z'):
                stayInAlphabet -= 26
            finalLetter = chr(stayInAlphabet)
        cipherText = ""
        cipherText += finalLetter

    print ("Your ciphertext is: ", cipherText)

    return cipherText

caesar(plainText, shift)
```

```python
def generate_cipher_text(text):
    cipher_text = ""
    for char in text:
        if char.isalpha():
            if char.islower():
                p = ord(char) - ord('a') + 1
                cipher_value = (7 * p + 6) % 26
                cipher_char = chr(cipher_value + ord('a') - 1)
            elif char.isupper():
                p = ord(char) - ord('A') + 1
                cipher_value = (7 * p + 6) % 26
                cipher_char = chr(cipher_value + ord('A') - 1)
        else:
            cipher_char = char
        cipher_text += cipher_char
    return cipher_text

# Test the function
text = "a"
cipher_text = generate_cipher_text(text)
print(cipher_text)
```

```python
def generate_cipher_text(text):
    cipher_text = ""
    for char in text:
        if char.isalpha():
            p = ord(char.lower()) - ord('a') + 1
            cipher_value = (3 * p - 4) % 26
            cipher_char = chr(cipher_value + ord('a'))
        else:
            cipher_char = char
        cipher_text += cipher_char
    return cipher_text

# Test the function
text = "andletgo"
cipher_text = generate_cipher_text(text)
print(cipher_text)
```
