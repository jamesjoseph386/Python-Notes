### The while Loop: can execute a set of statements as long as a condition is true.

```python
i = 1
while i <= 10:
  print(i, end=" ")
  i += 1

print('\n',"the last value of i is ",'\n',i)
```

```python
i = 1
while i < 10:
  print(i)
  if i == 3: 
    break  ### can stop the loop even if the while condition is true
  i += 1
```

```python
i = 0
while i < 10:
  i += 1
  if i == 3 or i == 6:
    continue    ####can stop the current iteration, and continue with the next
  print(i)
```
