## Write a Python program that takes a user input ( positive integer ). The program should then iterate through numbers from 1 to that number (both inclusive) and print "tik" if the current number is divisible by 3; otherwise, it should print "tok". 

```python
p = int(input("Enter the number:"))

for i in range(1, p + 1):
    if i % 3 == 0:
        print("tik")
    else:
        print("tok")
```
