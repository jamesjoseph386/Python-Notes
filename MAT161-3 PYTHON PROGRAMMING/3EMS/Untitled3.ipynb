# 1. Given a list of integers, write a program to compute the frequency of each integer and store the results in a dictionary.
# 2. Given a list of sentences, count the occurrences of each word across all sentences. Store the counts in a dictionary and print the dictionary.
# 3. Given a paragraph (a string), split it into words, count the frequency of each word, and store the counts in a dictionary. Print the dictionary showing each word and its frequency.

```python
#q1 
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4,5]

frequency_dict = {}

for number in numbers:
    if number in frequency_dict:
        frequency_dict[number] += 1
    else:
        frequency_dict[number] = 1

print(frequency_dict)
```

```python
#q2
sentences = ["hello world", "world of python", "hello python"]
frequency_dict = {}
for sentence in sentences:
    words = sentence.split()
    for word in words:
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            frequency_dict[word] = 1
print(frequency_dict) 
```

```python
#q3
paragraph = "This is a test. This test is only a test."
words1 = paragraph.lower()
words2=words1.replace('.', '')
words=words2.split()
frequency_dict = {}
for word in words:
    if word in frequency_dict:
        frequency_dict[word] += 1
    else:
        frequency_dict[word] = 1
print(frequency_dict)
```

```python
#q3
paragraph = "Can you believe that this will really work? Can you believe that this will really work? I mean, really, it’s hard to imagine!"  
 
words1 = paragraph.lower()  
 
punctuation = ['.', '?', '!', ',', ';', ':']   
for symbol in punctuation:  
    words1 = words1.replace(symbol, '')  

words = words1.split()  

frequency_dict = {}  
for word in words:  
    frequency_dict[word] = frequency_dict.get(word, 0) + 1  

print(frequency_dict) 
```
