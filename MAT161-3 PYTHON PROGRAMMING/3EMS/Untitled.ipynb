## Q6 Generate a dictionary where the keys are numbers and the values are Boolean indicating if the number is prime. 

```python
def is_prime(n):
    n=int(n)
    if n <= 1:  
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

list1 = inp_list()
dec = {}

for i in list1:
    if is_prime(i):
        dec[i] = True
    else:
        dec[i] = False

print(dec)
```

## Q Given a list of integers, write a program to compute the frequency of each integer and store the results in a dictionary.

```python
def is_tuple(item):
    return type(item) == tuple

def flatten(tup):
    flat_lst = []
    for item in tup:
        if is_tuple(item):
            for i in range(len(item)):
                flat_lst.append(item[i])  
        else:
            flat_lst.append(item)  
    return list(flat_lst)  

lst = [1,2,(3,(4,5)),6]
flat_list = flatten(lst)  
print(flat_list)  
```
