#### IF statements

```python
from random import randint

num = randint(1,10)
guess = eval(input('Enter your guess: '))
if guess==num:
    print('You got it!')
```

```python
from random import *

num = randint(1,10)
guess = eval(input('Enter your guess: '))
if guess==num:
    print('You got it!')
else:
    print('Sorry. The number is ', num)
```

## Conditional operators 

( and, or, not)

```python
grade=eval(input('Enter:'))
if grade>=80 and grade<90:
     print('Your grade is B.')
```

```python
score=eval(input('Enter:'))
time=eval(input('Enter:'))
if score>1000 or time>20:
     print('Game over.')
```

```python
score=eval(input('Enter:'))
time=eval(input('Enter:'))
if not (score<1000 or time<20):
    print('Game continues.')
```

###  elif

```python
 grade = eval(input('Enter your score: '))
if grade>=90:
    print('A')
elif grade>=80:
    print('B')
elif grade>=70:
    print('C')
elif grade>=60:
    print('D')
else:
    print('F')
```
