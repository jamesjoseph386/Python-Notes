## Rank of a Matrix

$A = \begin{bmatrix}1 & 7 & 5 & 6 \\ 7 & 2 & 6 & 7 \\-1 & -3 &-3 & 4\end{bmatrix}$</b>

```python
import numpy as np
A = np.array(([1,7,5,6],[7,2,6,7],[-1,-3,-3,4]))
print("A=",A) 
print("\n The rank of the matrix A is",np.linalg.matrix_rank(A))
```

$A = \begin{bmatrix}1 & 2 & 3\\ 2 & 1 & 4\\3 & 0 & 5\end{bmatrix}$</b>

```python
import numpy as np
A = np.array(([1,2,3],[2,1,4],[3,0,5]))
print("A=",A)
print("\n The rank of the matrix A is",np.linalg.matrix_rank(A))
```

## Properties

If A matrix is of order mXn, then $\rho$(A) $\le$ min{m,n} = minimum of m,n.

If A is of order nXn and |A| $\neq$ 0, then the rank of A = n.

If A is of order nXn and |A| = 0, then the rank of A will be less than n

```python
#Property1Example
import numpy as np
A = np.array(([4,2],[5,7],[8,9],[9,0]))
print("A=",A)
print("\n The rank of the matrix A is",np.linalg.matrix_rank(A))
```

```python
import numpy as np
A = np.matrix([[1,2,-1],[2,1,4],[3,3,4]])
B = np.matrix([[1],[2],[1]])
np.linalg.solve(A,B)
```

Q. Check whether the following system of homogeneous linear equation has non-trivial solution.<br>
$x_1 + 2x_2 - x_3 = 0$<br>
$2x_1 + x_2 +4x_3 = 0$<br>
$3x_1 + 3x_2 + 4x_3 = 0$

```python
import numpy as np
A1 = np.matrix([[1,2,3],[4,5,6],[7,8,9]])
B1 = np.matrix([[0],[0],[0]])
r = np.linalg.matrix_rank(A1)
n = A1.shape[1]
if (r==n):
    print("System has trivial solution")
else:
    print("System has", n-r, "non-trivial solution(s)")
```

```python
np.linalg.solve(A1,B1)
```

### System of Non-Homogeneous Linear Equations

Q. Examine the consistency of the following system of equations and solve if consistent<br>
$x_1 + 2x_2 - x_3 = 1$<br>
$2x_1 + x_2 + 4x_3 = 2$<br>
$3x_1 + 3x_2 + 4x_3 = 1$

```python
import numpy as np
A4 = np.matrix([[1,2,-1],[2,1,4],[3,3,4]])
B4 = np.matrix([[1],[2],[1]])
AB4 = np.concatenate((A4,B4),axis=1)
rA4 = np.linalg.matrix_rank(A4)
rAB4 = np.linalg.matrix_rank(AB4)
n = A4.shape[1]
if (rA4 <= rAB4):
    if(rA4==n):
        print("The system is consistent and has unique solution")
        print(np.linalg.solve(A4,B4))
    else:
        print("The system is consistent and has infinitely many solutions")
else:
    print("The sytem of equations is inconsistent and does not have any solution")
```
