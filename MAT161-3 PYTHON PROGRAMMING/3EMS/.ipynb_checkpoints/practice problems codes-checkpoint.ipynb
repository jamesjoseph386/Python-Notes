```python
#Q1
def intersection_of_sets(sets_list):
    # Check if the list of sets is empty
    if not sets_list:
        return set()
    
    # Initialize the intersection set with the first set in the list
    intersection_set = sets_list[0]
    
    # Iterate over the remaining sets and update the intersection set
    for current_set in sets_list[1:]:
        intersection_set &= current_set
    
    return intersection_set

# Example usage
set1 = {(1, 2), (3, 4), (5, 6)}
set2 = {(3, 4), (5, 6), (7, 8)}
set3 = {(5, 6), (7, 8), (3, 4)}

sets_list = [set1, set2, set3]
result = intersection_of_sets(sets_list)
print(result)  # Output: {(5, 6), (3, 4)}
```

```python
#Q2
# Step 1: Define the Lists
list1 = [1, 2, 3, 4]
list2 = ['a', 'b']

# Step 2: Determine the Maximum Length
max_length = max(len(list1), len(list2))

# Step 3: Pair Elements
paired_list = [(list1[i] if i < len(list1) else None,
                list2[i] if i < len(list2) else None) 
                for i in range(max_length)]

# Step 4: Output the Result
print(paired_list)
```

```python
#Q3
sets = [{1, 2, 3}, {2, 3, 4}, {4, 5, 6}]
sym_diff = sets[0]
for s in sets[1:]:
    sym_diff ^= s
print(sym_diff)  # {1, 5, 6}
```

```python
#Q4
set1 = {1, 2}
set2 = {'a', 'b'}
cartesian_product = {(a, b) for a in set1 for b in set2}
print(cartesian_product)  # {(1, 'a'), (1, 'b'), (2, 'a'), (2, 'b')}
```

```python
#Q5
nested_tuple = (1, (2, (3, (4, 5))), 6)
def flatten(tup):
    flat_list = []
    for item in tup:
        if isinstance(item, tuple):
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)
    return tuple(flat_list)
flat_tuple = flatten(nested_tuple)
print(flat_tuple)  # (1, 2, 3, 4, 5, 6)
```

```python
#Q6
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

numbers = range(1, 21)
prime_dict = {num: is_prime(num) for num in numbers}
print(prime_dict)  # {1: False, 2: True, 3: True, 4: False, 5: True, 6: False, 7: True, 8: False, 9: False, 10: False, 11: True, 12: False, 13: True, 14: False, 15: False, 16: False, 17: True, 18: False, 19: True, 20: False}
```

```python
#Q7
# Step 1: Define the List of Numbers
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# Step 2: Initialize an Empty Dictionary
frequency_dict = {}

# Step 3: Iterate Through the List and Update the Dictionary
for number in numbers:
    if number in frequency_dict:
        frequency_dict[number] += 1
    else:
        frequency_dict[number] = 1

# Step 4: Output the Frequency Distribution
print(frequency_dict)  # {1: 1, 2: 2, 3: 3, 4: 4}
```

```python
#Q8
sentences = ["hello world", "world of python", "hello python"]
frequency_dict = {}
for sentence in sentences:
    words = sentence.split()
    for word in words:
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            frequency_dict[word] = 1
print(frequency_dict)  # {'hello': 2, 'world': 2, 'of': 1, 'python': 2}
```

```python
#Q9
paragraph = "This is a test. This test is only a test."
words = paragraph.lower().replace('.', '').split()
frequency_dict = {}
for word in words:
    if word in frequency_dict:
        frequency_dict[word] += 1
    else:
        frequency_dict[word] = 1
print(frequency_dict)  # {'this': 3, 'is': 3, 'a': 3, 'test': 3, 'only': 1}
```

```python
#Q10
numbers = [123, 145, 234, 345, 412, 543]
grouped_by_first_digit = {}
for number in numbers:
    first_digit = str(number)[0]
    if first_digit not in grouped_by_first_digit:
        grouped_by_first_digit[first_digit] = []
    grouped_by_first_digit[first_digit].append(number)
print(grouped_by_first_digit)  # {'1': [123, 145], '2': [234], '3': [345], '4': [412], '5': [543]}
```
