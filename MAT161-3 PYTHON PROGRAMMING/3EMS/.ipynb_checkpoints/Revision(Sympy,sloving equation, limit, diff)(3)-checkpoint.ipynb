Q. Write a program which finds roots of the Quadratic equation $x^2+x+1$.

```python
import sympy as sp
x=sp.Symbol('x')
eqn=x**2+x+1
roots=sp.solve(eqn,x)
roots
```

Q. Write a program which finds roots of the Quadratic equation $x^3-6x^2+11x-6$

```python
import sympy as sp
x = sp.symbols('x')
equation = x**3 - 6*x**2 + 11*x - 6
roots = sp.solve(equation, x)
print("Roots:", roots)
```

Q. Write a program which finds roots of any Quadratic equation.

```python
import cmath

def find_roots(a, b, c):
    val = cmath.sqrt(b**2 - 4*a*c)
    root1 = (-b + val) / (2*a)
    root2 = (-b - val) / (2*a)

    return root1, root2

if __name__ == "__main__":
    a = float(input("Enter the coefficient a: "))
    b = float(input("Enter the coefficient b: "))
    c = float(input("Enter the coefficient c: "))
    roots = find_roots(a, b, c)
    print("Root 1:", roots[0])
    print("Root 2:", roots[1])
```

# Arc length:
## Arc length is the distance between 2points along a section of a curve. Determining the length of an irregular arc segment is also called rectification of a curve. The advent of infinitesimal calculus led to general formula that provides closed-form solutions in some cases.

## To compute the length of a curve on the interval [a,b] we compute the $\int\sqrt{1+(f'(x))^2}dx$

Q1: Find the arc length of $f(x)=x^{3/2}$ on $[0,2]$

```python
from sympy import *
x=Symbol("x")
f=diff(x**(3/2),x)
integrate(sqrt(1+f**2),(x,0,2))
```

Q2: Find the arc length of $f(x)=x$ on $[1,2]$

```python
from sympy import *
x=Symbol("x")
f=diff(x,x,1)
integrate(sqrt(1+f**2),(x,1,2))
```

Q3: Find the arc length of $f(x)= (1/3) (x^2  + 2)^{3/2}$ on $[0,1]$

```python
from sympy import *
x,y,z=symbols("x,y,z")
integrate(x*y*z,(z,0,1),(y,0,2),(x,0,3))
```
