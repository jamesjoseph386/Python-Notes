# Arc length:
## Arc length is the distance between 2points along a section of a curve. Determining the length of an irregular arc segment is also called rectification of a curve. The advent of infinitesimal calculus led to general formula that provides closed-form solutions in some cases.

## To compute the length of a curve on the interval [a,b] we compute the $\int\sqrt{1+(f'(x))^2}dx$

Q1: Find the arc length of $f(x)=x^{3/2}$ on $[0,2]$

```python
from sympy import *
x=Symbol("x")
f=diff(x**(3/2),x)
integrate(sqrt(1+f**2),(x,0,2))
```

Q2: Find the arc length of $f(x)=x$ on $[1,2]$

```python
from sympy import *
x=Symbol("x")
f=diff(x,x,1)
integrate(sqrt(1+f**2),(x,1,2))
```

Q3: Find the arc length of $f(x)= (1/3) (x^2  + 2)^{3/2}$ on $[0,1]$

```python
from sympy import *
x,y,z=symbols("x,y,z")
integrate(x*y*z,(z,0,x+y),(y,2,x),(x,0,3))
```
