### How do we user input two values and do the binary operations?

```python
# Get input from the user and convert to integers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Perform operations
sum_result = num1 + num2
difference_result = num1 - num2
product_result = num1 * num2

# Handle division by zero
if num2 != 0:
    division_result = num1 / num2
else:
    division_result = "Cannot divide by zero"

# Display results
print("Sum:", sum_result)
print("Difference:", difference_result)
print("Product:", product_result)
print("Division:", division_result)
```

### Inbuilt Mathematical Functions

| Function            | Description      |
| ------------------- | ---------------- |
| `math.sqrt(x)`      | Square root      |
| `math.pow(x, y)`    | x to the power y |
| `math.factorial(x)` | Factorial        |
| `math.pi`           | π constant       |


```python
import math
print(math.sqrt(25))      # Square root
print(math.pow(2, 3))     # Power
print(math.factorial(5))  # Factorial
print(math.pi)            # Pi constant
```

### Write a python program to find the area of a circle with radius 7.

```python
import math
radius = 7
area = math.pi * radius ** 2

print("Area of circle:", area)
```
