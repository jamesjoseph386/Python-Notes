# Harmonic Function

A function $u(x,y)$ is called harmonic if it is twice continuously differentiable and satisfies the following partial
differential equation: $\frac{d^2u}{dx^2} + \frac{d^2u}{dy^2}= 0$

### Aim

To check if the function is harmonic and if it is harmonic function,and to check if a function 'v' is the harmonic
conjugate of the harmonic function 'u'.

### Check if function is harmonic

```python
from sympy import *
init_printing(use_latex=True)
def harmonic(q):
    print("The function no. is: ",q)
    dqdx = diff(q,x,2)
    dqdy = diff(q,y,2)
    if dqdx+dqdy==0:
        print('This is a harmonic function.')
    else:
        print('This is not a harmonic function.')
```

```python
x,y=symbols('x,y')
```

### Q1. $x^3-3xy^2$

```python
harmonic(x**3-3*x*y**2)
```

### Q2. $x^2$-$y^2$+x+1

```python
harmonic(x**2-y**2+x+1)
```

### Q3. $\frac{1}{2}cos(x^2+y^2)$

```python
harmonic((1/2)*cos(x**2+y**2))
```

### To check if a function 'v' is the harmonic conjugate of the harmonic function 'u'

```python
def verify(u,v):
    x,y=symbols('x,y')
    dv_dx2=diff(v,x,2)
    dv_dy2=diff(v,y,2)
    du_dx=diff(u,x)
    du_dy=diff(u,y)
    dv_dx=diff(v,x)
    dv_dy=diff(v,y)
    if dv_dx2 + dv_dy2==0:
        if du_dx==dv_dy and du_dy==-dv_dx:
            print("v is a harmonic conjugate of u")
        else:
            print("CR Equations are not satisfied")
    else:
        print("v itself is not a harmonic function")
```

### Q1. Check if $cos(x)sinh(y)$ is a harmonic conjugate of $sin(x)cosh(y)$

```python
verify(sin(x)*cosh(y),cos(x)*sinh(y))
```

### Q2. Check if $\frac{-x^2}{2}+\frac{y^2}{2}$ is a harmonic conjugate of $x^2-y^2+x+1$

```python
verify(x**2-y**2+x+1,-x**2/2+y**2/2)
```

### Q3. Check if $\frac{-x^2}{2}+\frac{y^2}{2}$ is a harmonic conjugate of $xy$

```python
verify(x*y,-x**2/2+y**2/2)
```
