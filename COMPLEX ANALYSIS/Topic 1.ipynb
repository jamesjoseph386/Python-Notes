## 3) Complex Functions in Python

#### Python converts the real numbers x and y into complex using: the function complex(x,y)
#### The real part can be accessed using: the function real() and imaginary part can be represented by: imag().

```python
# Initializing real numbers
x = 7
y = 14
# converting x and y into complex number
z = complex(x,y) # z = x+iy
z1=complex(8,10)
print("The complex nos. are", z,"&",z1)
```

#### Extracting Real and Imaginary Part from a Complex Number

```python
z2=complex(-1,5)
print("The complex no. is", z2)

# printing real and imaginary part of complex number
print ("The real part of complex number is : ",end="")
print (z2.real)

print ("The imaginary part of complex number is : ",end="")
print (z2.imag)
```

```python
p=(float(input("Enter real part of the complex number:")))
q=(float(input("Enter imaginary part of the complex number:")))

print("")

z4=complex(p,q)

print("The complex nos. is", z4)
print("")

# the complex conjugate of a complex number is the number with an equal real part and an imaginary part equal in magnitude,
# but opposite in sign

print ("The conjugate of the given complex number is : ",end="")
print (z4.conjugate())
```

#### Polar and rectangular form of the complex number

1). The polar form of a complex number is another way to represent a complex number.<br>
2). The form $z=a+bi$ is called the rectangular coordinate form of a complex number.<br><br>

#### The polar form of a complex number $z=a+bi$ is $z=r(cosθ+isinθ)$ , where $r=|z|=\sqrt{a^2+b^2}$ , $a=r~cos~θ$  and $ b=r~sin~θ$.

##### Q. Find the polar and rectangular form of the complex number

```python
z7=8-5j
print("The complex no. is", z7)
print("")
# converting complex number into polar using polar()
a1=polar(z7)
print("Polar form is:", a1)
print("")
# converting complex number into rectangular using rect()
a2=rect(a1[0],a1[1])
print("Rectangular form is:", a2)
```
