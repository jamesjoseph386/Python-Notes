# Lab 7-Milne Thomson Method: To find f(z) if u+v is given

The Milne-Thomson method is a method of finding a holomorphic function, whose real or imaginary part is
given. The method greatly simplifies the process of finding the holomorphic function whose real or imaginary
part is given. It is named after Louis Melville Milne-Thomson.

A holomorphic function is a complex-valued function of one or more complex variables that is, at every point of its domain, complex differentiable in a neighborhood of the point.

We need to obtain the analytic function f(z) when the real and imaginary part of the function is given using the Milney-
Thompson method. Here $u$ is the real part and $v$ is the imaginary part. The function $u+v$ is given, we need to find $u$ and $v$ by partial differentiation and some simplifications. Once we get $u(x,y)$ and $v(x,y)$, $f(z)=u(z,0)+i v(z,0)$ i.e we put x=z and y=0.

### Find the analytic function f(z)

Q1. $u + v = e^x(cosy + siny)$

```python
import sympy
from math import *
x,y,z = sympy.symbols("x,y,z")
#g=exp(x)*(cos(y)+sin(y))
g = sympy.exp(x)*(sympy.cos(y)+sympy.sin(y))
U = sympy.diff(g,x) #e^x(cosy+siny)
V = sympy.diff(g,y) #e^x(-siny+cosy)
dudx = (U+V)/2
dvdx = (U-V)/2
Z = dudx + 1j*dvdx
Z1 = Z.subs({x:z,y:0})
F = sympy.integrate(Z1,z)
print(F)
```

Q1. $u + v = e^x(cosy + siny)$

```python
from math import *
import sympy as sp
x,y,z = symbols("x,y,z")
f = sp.exp(x)*(sp.cos(y)+sp.sin(y))
U = diff(f,x)
V = diff(f,y)
dudx = (U+V)/2
dvdx = (U-V)/2
Z = dudx + 1j*dvdx
Z1 = Z.subs({x:z,y:0})
F = integrate(Z1,z)
print(F)
```
