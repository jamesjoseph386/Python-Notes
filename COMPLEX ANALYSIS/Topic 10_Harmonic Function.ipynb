# To find the harmonic conjugate v of a harmonic function

### Harmonic Conjugate

The harmonic conjugate to a given function $u(x,y)$ is a function $v(x,y)$ such that: $f(x,y)=u(x,y)+iv(x,y)$ is
complex differentiable (i.e., satisfies the Cauchy-Riemann equations). It is given by:$$v(z)=\int_{z_0}^{z}U_x \,dy - U_y \,dx + C$$ where $U_x=\frac{du}{dx}, U_y=\frac{du}{dy}$ and $C$ is a constant of integration.

### AIM
To find the harmonic conjugate of any given harmonic conjugate (post verifying the same).

```python
from numpy import *
from sympy import *
import math

def conjugate():
    init_printing(use_latex=True)
    x,y,c=symbols("x,y,c")
    f=input("Enter the function u:")
    print("")
    dx2=diff(f,x,2)
    dy2=diff(f,y,2)
    if dx2+dy2==0:
        print("The function ",f," is harmonic.")
    else:
        print("The function ",f," is not harmonic.")
    dvx=diff(f,x)
    dvy=diff(f,y)
    ivx=integrate(dvx,y)
    ivy=(-integrate(dvy,x))
    v=ivx+ivy
    print("The harmonic conjugate of u is:")
    display(v)
```
