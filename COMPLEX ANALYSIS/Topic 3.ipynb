# $n^{th}$ roots of a complex number

#### Q1) Find the 4th roots of the complex number $z= 5(cos(\pi/3) + i sin(\pi/3))$

```python
from cmath import *
from math import *
z=5*(complex(cos(pi/3),sin(pi/3)))
print("The complex number is",z)
for k in range(4):
    root=(5**(1/4))*(complex(cos(((pi/3)+(2*k*pi))/4),(sin(((pi/3)+(2*k*pi))/4))))
    print("Root",k+1,"is",root)
```

## Q2) Find roots of $z=(2i)$

```python
a=complex(0,2)
print("The complex number is",a)
r,theta=polar(a)
n=2
for k in range(n):
    root=(r**(1/n))*complex(cos((theta+(2*k*pi))/n),sin((theta+(2*k*pi))/n))
    print("Root",k+1,"of the complex number",z,"is",root)
```

# 1) $(1+i)^\frac{1}{5}$

```python
import matplotlib.pyplot as plt
b=complex(1,1)
print("The given complex number is",b)
r,theta=polar(b)
n=5
rts=[]
for k in range(n):
    root=(r**(1/n))*complex(cos((theta+(2*k*pi))/n),sin((theta+(2*k*pi))/n))
    print("Root",k+1,"is",root)
    rts.append(root)
for i in range(len(rts)):
    plt.scatter(rts[i].real,rts[i].imag,marker="*")
    
plt.text(rts[i].real,rts[i].imag,b)
plt.show()
```

# 2) $(3+4i)^\frac{1}{2}$

```python
c=complex(3,4)
print("The given complex number is",c)
r,theta=polar(c)
n=2
rts=[]
for k in range(n):
    root=(r**(1/n))*complex(cos((theta+(2*k*pi))/n),sin((theta+(2*k*pi))/n))
    print("Root",k+1,"is",root)
    rts.append(root)
for i in range(len(rts)):
    plt.polar(phase(rts[i]),abs(rts[i]),marker="*")
```
