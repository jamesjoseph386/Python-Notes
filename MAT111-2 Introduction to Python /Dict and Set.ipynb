### An Introduction to Dictionary in Python:

#### $\bullet$ Dictionaries are used to store data values in key:value pairs. (unordered, mutable, key values should be unique)
#### $\bullet$ A dictionary is a collection which is ordered*, changeable and do not allow duplicates.
#### $\bullet$ Dictionaries are written with curly brackets, and have keys and values:

Syntax: dictionary = {key: value, key : value , ... ,key: value}

#### Dictionary Items:
#### $\bullet$ Dictionary items are ordered, changeable, and does not allow duplicates.
#### $\bullet$ Dictionary items are presented in key:value pairs, and can be referred to by using the key name.
#### $\bullet$ The values in dictionary items can be of any data type.

```python
#Create and print a dictionary:
person = {
    'first_name': 'John',
    'last_name': 'Doe',
    'age': 25,
    'favorite_colors': ['blue', 'green'],
    'active': True
}
print("Details of that Person:\n",person)
```

```python
person2 = {"name": "Jessa", "country": "USA", "telephone": 1178}
print(person2)
```

```python
emptydict = {}
print(emptydict) 
```

```python
print(type(person))
```

```python
print(person2.keys()) ## keys()- Returns the list of all keys present in the dictionary.
```

```python
print(person2.values()) ### values() - Returns the list of all values present in the dictionary
```

```python
#items()-Returns all the items present in the dictionary.
#Each item will be inside a tuple as a key-value pair.
print(person2.items())
```

#### Accessing values in a Dictionary:
##### To access a value by key from a dictionary, place the key inside square brackets: dict[key]

```python
print(person['first_name'])
```

```python
print(person['favorite_colors'])
```

```python
print(person.get('last_name')) #### by using get()
```

### get() - If the key doesn’t exist, the get() method returns None instead of throwing a KeyError.

```python
### for instance
print(person['father name'])
```

```python
print(person.get('father name'))
```

### The get() method also returns a default value when the key doesn’t exist by passing the default value to its second argument.

```python
print(person.get('father name', 'does not exsits'))
```

### Adding new key-value pairs

```python
person['gender'] = 'male'
print("The updated dictionary is: \n", person)
```

```python
## We can also add more than one key using the update() method.
```

```python
person.update({"weight": 50, "height": 6})
print("The updated dictionary is: \n", person)
```

### Modifying values in a key-value pair : Syntax: dict[key] = new_value


```python
person['age']=26
print("The updated dictionary is: \n", person)
```

#### To remove a key-value pair by a key, you use the del statement: del dict[key]

```python
del person['active']
print(person)
```

### Create a set from a list

```python
number_list = [20, 30, 20, 30, 50, 30]
# create a set from a list
sample_set2 = set(number_list)
print(sample_set2)
```

```python
## The items of the set are unordered and they don’t have any index number.
book_set = {"Harry Potter", "Angels and Demons", "Atlas Shrugged"}
for book in book_set:
    print(book)
```

### The $in$ operator checks whether the item is present in the set, and returns True if it present otherwise, it will return False.

```python
book_set = {"Harry Potter", "Angels and Demons", "Atlas Shrugged"}
if 'Harry Potter' in book_set:
    print("Book exists in the book set")
else:
    print("Book doesn't exist in the book set")
```

```python
# check another item which is not present inside a set
print("A Man called Ove" in book_set)  
```

```python
print(len(book_set))
```

### We can add new items to the set using the following two ways.

```python
book_set = {'Harry Potter', 'Angels and Demons'}
# add() method
book_set.add('The God of Small Things')
# display the updated set
print(book_set)
```

```python
# update() method to add more than one item
book_set.update(['Atlas Shrugged', 'Ulysses'])
# display the updated set
print(book_set)
```

```python
color_set = {'red', 'orange', 'yellow', 'white', 'black'}
# remove single item
color_set.remove('yellow')
print(color_set)
```

```python
# remove all items
color_set.clear()
print(color_set)
```

### Set Operations

Union -> | or union()

Intersection -> & or intersection()

Difference -> -	or difference()

Symmetric Difference -> ^ or symmetric_difference()

```python
color_set = {'violet', 'indigo', 'blue', 'green', 'yellow'}
remaining_colors = {'indigo', 'orange', 'red'}

# union of two set using OR operator
new_color_set = color_set | remaining_colors
print("The union of two sets:",new_color_set)


# union using union() method
new_colors = color_set.union(remaining_colors)
print("The union of two sets:",new_colors)
```

```python
new_color_set = color_set & remaining_colors
print("The intersection of two sets:",new_color_set)
```

```python
new_color_set = color_set - remaining_colors
print("The difference:",new_color_set)
```

```python
# symmetric difference
unique_items = color_set ^ remaining_colors
print("The symmetric difference:",unique_items)
```
