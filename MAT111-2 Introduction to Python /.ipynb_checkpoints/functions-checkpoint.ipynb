### Function
#### A function is a block of code that perform a specific task Types of function:
#### Standard library function- in-built functions (eg. $print(), sqrt(), pow(),$ etc)

#### (These library fns are defined inside the module, eg is defined inside the 'math' module)

#### User-defined function- we can make our own function

```python
### Built-in Math Functions
```

```python
x = min(3, 1, 2)
y = max(5, 2, 6)
print(x)
print(y)
```

```python
x = pow(2, 8)
print(x)
```

```python
import math
x = math.ceil(1.4)
y = math.floor(1.4)
print(x) # returns 2
print(y) # returns 1
```

```python
import math
square_root= math.sqrt(4)
print('square root of 4 is:', square_root)
power= pow(2,3)
print('2 to the power 3 is:', power)
```

```python
import math
x = math.pi
print(x)
```

### Python function arguments
#### A function can also have arguments. An argument is a value that is accepted by a function.

```python
## A function with 2 arguments
def add_num(num1, num2):
    sum=num1+num2
    print('Sum:', sum) #### Add the corresponding values while calling the function
add_num(20,40)
```

### Python Variables
#### 3 types of variables:
#### Local variable
#### Global variable

```python
def greet():
## Local variable
    message='Hello, I am a local variable'
    print('Local:', message)
greet()
```

```python
def greet():
## Local variable
    message='Hello, I am a local variable'
    print('Local:', message)
greet()

def welcome():
    message = 'hey'
    print('Local:',message)
welcome()
```

```python
# declare global variable
msg= 'Helo, I am a global variable'
def greet():
    print('Local:', msg)
greet()

def welcome():
    msg='hi'
print('Global:', msg)
```
